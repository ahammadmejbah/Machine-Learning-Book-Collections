## Predict a binomial variable, means only two possible outcomes

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
```

```python
# import iris dataset
iris = datasets.load_iris()
```

```python
# split data featurs and labels
features = iris['data']
labels = iris['target']
```

```python
# slice data - select sepal_length
X = features[:,:1]
# classify target according to virginica, if virinica = 1 otherwise 0 
y = (labels == 2).astype(np.int8)
```

```python
y
```

```python
# create model
clf = LogisticRegression()
clf.fit(X,y)
```

```python
predict = clf.predict([[9.1]])
```

```python
predict
```

```python
# calculate coeficient and intercept
weight= clf.coef_
bias =  clf.intercept_
```

```python
# find probability
# create a fucntion
def proba(clf, x):
    log_odds = weights * x + bias
    odds = np.exp(log_odds)
    prob = odds/(1+odds)
    return prob
```

```python
# 99% probabiltiy, its virginica
proba(clf, ([[9.1]]))
```
